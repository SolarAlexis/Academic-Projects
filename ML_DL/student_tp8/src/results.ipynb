```python
import os

import torch
from tensorboard.backend.event_processing import event_accumulator

import matplotlib.pyplot as plt
```

```python
script_dir = os.getcwd() 
runs_dir = os.path.join(script_dir, "log_exo1")

ea = event_accumulator.EventAccumulator(runs_dir)
ea.Reload()

# Rcupre les valeurs de train_loss
values1 = ea.Scalars("train_loss")

# Extraire les itrations et les valeurs de f(beta)
epochs = [entry.step for entry in values1]
train_loss = [entry.value for entry in values1]

# Rcupre les valeurs de test_loss
values2 = ea.Scalars("test_loss")

epochs2 = [entry.step for entry in values2]
test_loss = [entry.value for entry in values2]

# Rcupre les valeurs de test_loss
values3 = ea.Scalars("train_accuracy")

epochs3 = [entry.step for entry in values3]
train_accuracy = [entry.value for entry in values3]

# Rcupre les valeurs de test_loss
values4 = ea.Scalars("test_accuracy")

epochs4 = [entry.step for entry in values4]
test_accuracy = [entry.value for entry in values4]

# plot
plt.figure(figsize=(9, 6))
plt.plot(epochs, train_loss, label="train_loss")
plt.plot(epochs2, test_loss, label="test_loss")
plt.plot(epochs3, train_accuracy, label="train_accuracy")
plt.plot(epochs4, test_accuracy, label="test_accuracy")
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.title('loss over Epochs')
plt.legend()
plt.show()
```
