```python
import os

import matplotlib.pyplot as plt

from tensorboard.backend.event_processing import event_accumulator
```

```python
script_dir = os.getcwd() 
runs_dir = os.path.join(script_dir, "run_descente")

ea = event_accumulator.EventAccumulator(runs_dir)
ea.Reload()

# Récupére les valeurs de train_loss
values = ea.Scalars('Loss/train')

# Extraire les itérations et les valeurs de f(beta)
n_iter = [entry.step for entry in values]
train_loss = [entry.value for entry in values]

# plot
plt.figure(figsize=(9, 6))
plt.plot(n_iter, train_loss)
plt.xlabel('Iterations')
plt.ylabel('train_loss')
plt.title('train loss over iterations')
plt.show()
```
