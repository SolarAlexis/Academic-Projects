```python
# a) 

def Sn(n):
    res = 0
    for i in range(n+1):
        res = res + 1/((2*i+1)**2)
    return res
```

```python
# b)

def comparaison(epsilon):
    i = 0
    while abs(np.pi**2-8*Sn(i)) >= epsilon:
        i = i+1
    return i
```

```python
# a)

def  centre_rayon(a,b,c):
    x = a / (-2)
    y = b / (-2)
    centre = (x , y)
    rayon = math.sqrt(abs(c - x**2 - y**2))
    return centre , rayon
```

```python
# b)

def  plot_cercle(a,b,c):
    centre, rayon = centre_rayon(a,b,c)
    x = np.linspace(0,2*math.pi,50)
    y = centre[0] + rayon*np.cos(x)
    z = centre[1] + rayon*np.sin(x)
    plt.plot(y,z)
    plt.axis("equal")
    plt.plot(centre[0],centre[1], marker="o", color="blue")
```

```python
# b) cercle d'équation x² + y² = 1 <=> x² + y² + 0x + 0y - 1 = 0 
#                                  <=> a = 0, b = 0, c = -1

plot_cercle(0,0,-1)
```

```python
# b) cercle d'équation (x-1)² + y² = 1 <=> x² + y² - 2x + 0y + 0 = 0
#                                      <=> a = -2, b = 0, c = 0

plot_cercle(-2,0,0)
```

```python
# b) cercle d'équation x² + (y-1)² = 1 <=> x² + y² + 0x - 2y + 0 = 0
#                                      <=> a = 0, b = -2, c = 0

plot_cercle(0,-2,0)
```

```python
# c)

def cercle_par_points(X,Y):
    a = np.array([[X[0],Y[0],1],[X[1],Y[1],1],[X[2],Y[2],1]])
    b = np.array([-X[0]**2-Y[0]**2,-X[1]**2-Y[1]**2, -X[2]**2-Y[2]**2])
    x = np.linalg.solve(a, b)
    return centre_rayon(x[0],x[1],x[2])
```

```python
# c) on définit la fonction test1 pour dessiner le cercle à partir de a = x[0], b = x[1], c = x[2]

def test1(X,Y):
    a = np.array([[X[0],Y[0],1],[X[1],Y[1],1],[X[2],Y[2],1]])
    b = np.array([-X[0]**2-Y[0]**2,-X[1]**2-Y[1]**2, -X[2]**2-Y[2]**2])
    x = np.linalg.solve(a, b)
    plot_cercle(x[0],x[1],x[2])
```

```python
# c) on définit la fonction testcol pour s'assurer que nos 3 points ne sont pas colinéaires

def testcol(X,Y):
    if (X[0]*Y[1]-X[1]*Y[0] != 0) and (X[1]*Y[2]-X[2]*Y[1] != 0) and (X[0]*Y[2]-X[2]*Y[0] != 0):
        return False
    return True
```

```python
# d)

def cercle_moindres_carres(X,Y):
    longueur = len(X)
    a = np.array([[X[i],Y[i],1] for i in range(longueur)])
    b = np.array([-X[i]**2-Y[i]**2 for i in range(longueur)])
    x = np.linalg.lstsq(a, b, rcond=None)[0]
    return centre_rayon(x[0],x[1],x[2])
```

```python
# d) on définit la fonction test2 pour dessiner le cercle à nouveau

def test2(X,Y):
    longueur = len(X)
    a = np.array([[X[i],Y[i],1] for i in range(longueur)])
    b = np.array([-X[i]**2-Y[i]**2 for i in range(longueur)])
    x = np.linalg.lstsq(a, b, rcond=None)[0]
    plot_cercle(x[0],x[1],x[2])
    for i in range(longueur):
        plt.plot(X[i],Y[i], marker="o", color="red")
```

```python
# a)

def dft_complex(x):
    stemp = [0 for i in range(len(x))]
    # j'utilise une double somme indéxée sur i et j correspondant à k et n dans la formule
    for i in range(len(x)):
        a = 0j
        for j in range(len(x)):
            a = a + x[j]*np.exp(-2j*np.pi*i*j/len(x))
        stemp[i] = 1/math.sqrt(len(x))*a
    return np.array([stemp[i] for i in range(len(x))])
```

```python
# a) on vérifie à l'aide de la fonction np.fft.fft

x1 = np.random.random(4) + np.random.random(4) * 1j
dft_complex(x1), np.fft.fft(x1,norm="ortho")
```

```python
# b)

def idft_complex(s):
    xtemp = [0 for i in range(len(s))]
    # j'utilise une double somme indéxée sur i et j correspondant à n et k dans la formule
    for i in range(len(s)):
        a = 0 + 0j
        for j in range(len(s)):
            a = a + s[j]*np.exp(2j*np.pi*i*j/len(s))
        xtemp[i] = 1/math.sqrt(len(s))*a
    return np.array([xtemp[i] for i in range(len(s))])
```

```python
from scipy.integrate import quad

# e) On définit f1 et f2 pour en faire l'intégrale dans la fonction ft_real

def f1(x, n):
    return f(x)*np.cos(2*np.pi*n*x/T)

def f2(x, n):
    return f(x)*np.sin(2*np.pi*n*x/T)

def ft_real(f, T, N):
    a0 = 1/T*quad(func=f, a=-T/2, b=T/2, limit=500)[0]
    an = np.array([2/T*quad(func=f1, a=-T/2, b=T/2, args=i, limit=500)[0] for i in range(1,N+1)])
    bn = np.array([2/T*quad(func=f2, a=-T/2, b=T/2, args=i, limit=500)[0] for i in range(1,N+1)])
    return a0 , an, bn
```

```python
# e) On définit une fonction f avec une période T et un nombre de coefficients N, on peut alors calculer ft_real(f,T,N)

def f(x):
    return np.cos(x)
T = np.pi*2
N = 3

ft_real(f, T, N)
```

```python
# f) On définit ftemp pour appliquer la fonction aux valeurs dans l'intervalle [-1,1]

def ftemp(x):
    return np.exp(x) + np.exp(-x) + x**3 - x

def fonction(x):
    y = x%2
    for i in range(len(x)):
        if y[i] > 1:
            y[i] = y[i] - 2
        if y[i] < -1:
            y[i] = y[i] + 2
    return ftemp(y)
```

```python
# f) on trace le graphe de la fonction sur [-5,5]

x = np.linspace(-5,5,1000)
y = fonction(x)
plt.plot(x,y)
plt.show()
```

```python
# g) On réutilise les fonctions des questions e) et f) et on définie la fonction2 pour avoir abs(f)² puis la fonction3 pour
#    avoir une valeur de "fonction" à partir d'un tableau numpy de dimension (1,1)

T = 2
N = 100
def fonction2(x):
    return abs(np.exp(x) + np.exp(-x) + x**3 - x)**2

def fonction3(x):
    return fonction(np.array([x]))

# On calcule la partie gauche de l'équation
a, _, _ = ft_real(fonction2, T, N)
resultat1 = a

# Puis on calcule la partie droite de l'équation
b, c, d= ft_real(fonction3, T, N)
resultat2 = b**2+1/2*np.sum(c**2+d**2)
resultat1,resultat2
```
